```python
t=np.linspace(0,2*np.pi,20)
print(t)
```

```python
# Hacemos uso de las funciones universales como sen, cos, tan, e^x, etc
x=np.sin(t)
print(x)
```

```python
o=np.arcsin(x)
o
```

```python
# se grafica los 50 valores vs el seno de esos valores
# No se olvide que numpy realiza operaciones vectorizadas
plt.plot(t,x)
plt.grid()
plt.show()
```

```python
# Hacemos uso de otra función universal coseno(x)
y=np.cos(t)
print(y)
```

```python
# se grafica los 50 valores vs el seno de esos valores
# No se olvide que numpy realiza operaciones vectorizadas
plt.plot(t,y)
plt.grid()
plt.show()
```

## SUPERPOSICIÓN

```python
# Utilizamos el método figure para establecer el lienzo blanco
# al hacer uso del método plot 2 veces, se hará 2 gráficos en uno
# se tiene un solo lienzo.
plt.figure()
plt.plot(x)
plt.plot(y)
plt.grid()
plt.show()
```

## GRÁFICAS INDEPENDIENTES

```python
# Utilizamos el método figure para establecer el lienzo blanco sobre el cual
# trabajaremos utilizando el otro metodo plot
plt.figure()
plt.plot(y)
plt.grid()
plt.figure()
plt.plot(x)
plt.grid()
plt.show()
```

## GRÁFICA CON UN VERSUS

```python
r=np.linspace(0,2*np.pi,60)
y1=np.sin(r)
y2=np.sin(2*r)
plt.figure()
plt.plot(r,y1)
plt.plot(r,y2)
plt.grid()
plt.show()
```

## SUPERPOSICIÓN DE GRAFICAS CON VERSUS

```python
plt.figure()
plt.plot(r,y1,color='c',linewidth=5)
plt.plot(r,y2,'g--*',linewidth=2)
plt.legend(['sen(x)','sen(2x)'])
plt.show()
```

```python
#%matplotlib qt
plt.figure(figsize=(13,10))
s1=np.linspace(0,5,5)
s2=np.linspace(0,5,10)
s3=np.linspace(0,5,30)
t1=s1**3
t2=s2**3
t3=s3**3
plt.subplot(3,1,1)
plt.plot(s1,t1,'r')
plt.axhspan(40,50.6,color='#1f77b4',alpha=0.3)
plt.axvspan(2,3,color='pink',alpha=0.5)
plt.xlabel('EJE X')
plt.ylabel('EJE Y')
plt.title('FUNCIÓN CÚBICA CON 10 VALORES')
plt.subplot(3,1,2)
plt.plot(s2,t2,'g')
plt.axhline(40,c='black')
plt.axvline(3.4,c='black')
plt.xlabel('EJE X')
plt.ylabel('EJE Y')
plt.title('FUNCIÓN CÚBICA CON 30 VALORES')
plt.subplot(3,1,3)
plt.plot(s3,t3)
plt.text(2,80,'CURVA')
plt.xlabel('EJE X')
plt.ylabel('EJE Y')
plt.title('FUNCIÓN CÚBICA CON 50 VALORES')
plt.tight_layout()
plt.show()
```

## MANIPULANDO EL Data frames
Los data frames son estructuras de datos de dos dimensiones (rectangulares) que pueden contener datos de diferentes tipos, por lo tanto, son heterogéneas. Esta estructura de datos es la más usada para realizar análisis de datos y seguro te resultará familiar si has trabajado con otros paquetes estadísticos.

```python
# READING DATA IN PANADAS

# LEER CABECERAS O TÍTULOS:
print(df_csv.columns,'\n')
print('----------------------------------1-------------------------------------------','\n')

# LEER UNA COLUMNA EN ESPECÍFICA
print(df_csv['Name'],'\n') # print(df_csv.Name) -> 2da forma
print('----------------------------------2-------------------------------------------','\n')

# Y PODEMOS ESPECIFICAR DE ESA COLUMNA EN ESPECÍFICOCUANTO QUEREMOS LEER
print(df_csv.Name[2:5],'\n')
print('----------------------------------3-------------------------------------------','\n')

# TAMBIÉN PODEMOS OBTENER LAS COLUMNAS QUE QUERAMOS
print(df_csv[['Type 1','HP']],'\n')
print('----------------------------------4-------------------------------------------','\n')

# TAMBIÉN PODEMOS LEER FILAS Y CUANTAS QUEREMOS PERO DESDE EL PRINCIPIO
print(df_csv.head(4),'\n')             
print('----------------------------------5-------------------------------------------','\n')

# OBTENEMOS TAMBIÉN LAS FILAS QUE QUERAMOS
print(df_csv.iloc[2:4],'\n')
print('----------------------------------6-------------------------------------------','\n')

# PODEMOS ESCOJER UNA FILA EN PRTICULAR
print(df_csv.iloc[2])
print('----------------------------------7-----------------------------------------','\n')

# TMABIÉN PODEMOS ESCOGER LOS ÚLTIMOS DE LA LISTA
print(df_csv.tail(3))
print('----------------------------------8-----------------------------------------')
```
